# **[N-th Tribonacci Number](https://leetcode.com/explore/learn/card/dynamic-programming/631/strategy-for-solving-dp-problems/4041/)** 

### Recursion


```python
class Solution:
    def tribonacci(self, n: int) -> int:
        if n==0:
            return 0
        if n==1:
            return 1
        if n==2:
            return 1
        return self.tribonacci(n-1)+self.tribonacci(n-3)+self.tribonacci(n-2)

s= Solution()
```

```python
n = 4
# Output: 4
s.tribonacci(n)
```

```python
n = 25
# Output: 1389537
s.tribonacci(n)
```

### Memoization


```python
class Solution:
    def tribonacci(self, n: int, temp= True) -> int:
        if temp:
            self.t= [-1 for i in range(n+1)]
            temp= False

        if n==0:
            return 0
        if n==1:
            return 1
        if n==2:
            return 1
        
        if self.t[n]!=-1:
            return self.t[n]
        self.t[n]= self.tribonacci(n-1, temp)+self.tribonacci(n-3, temp)+self.tribonacci(n-2, temp)
        return self.t[n]

s= Solution()
```

```python
n = 4
# Output: 4
s.tribonacci(n)
```

```python
n = 25
# Output: 1389537
s.tribonacci(n)
```

### Bottom Up


```python
class Solution:
    def tribonacci(self, n: int, temp= True) -> int:
        if temp:
            self.t= [-1 for i in range(n+1)]
            temp= False

        if n==0:
            return 0
        if n==1:
            return 1
        if n==2:
            return 1

        self.t[0]=0
        self.t[1]=1
        self.t[2]=1
        for i in range(3, n+1):
            self.t[i]= self.t[i-1]+self.t[i-2]+self.t[i-3]
        return self.t[n]

s= Solution()
```

```python
n = 4
# Output: 4
s.tribonacci(n)
```

```python
n = 25
# Output: 1389537
s.tribonacci(n)
```
