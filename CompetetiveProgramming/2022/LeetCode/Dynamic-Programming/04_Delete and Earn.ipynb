### Recursion

```python
class Solution:
    def delete_and_earn(self, n):
        # Base Condition
        if n<=0:
            return 0

        delete_n= n*self.freq[n]
        if n>1:
            delete_n+= self.delete_and_earn(n-2)

        not_delete_n= self.delete_and_earn(n-1)
        return max(delete_n, not_delete_n)
        
    def deleteAndEarn(self, nums: list[int]) -> int:
        n= len(nums)
        max_element= max(nums)

        self.freq= [0 for i in range(max_element+1)]
        for i in nums:
            self.freq[i]+=1

        print('indx: ', [i for i in range(max_element+1)])
        print('freq: ', self.freq)
        return self.delete_and_earn(max_element)


s= Solution()
```

```python
nums = [3,4,2]
# Output: 6
s.deleteAndEarn(nums)
```

```python
nums = [2,2,3,3,3,4]
# Output: 9
s.deleteAndEarn(nums)
```

### Memoization


```python
class Solution:
    def delete_and_earn(self, n):
        # Base Condition
        if n<=0:
            return 0

        if self.t[n]!=-1:
            return self.t[n]

        delete_n= n*self.freq[n]
        if n>1:
            delete_n+= self.delete_and_earn(n-2)

        not_delete_n= self.delete_and_earn(n-1)
        self.t[n]= max(delete_n, not_delete_n)
        return self.t[n]
        
    def deleteAndEarn(self, nums: list[int]) -> int:
        n= len(nums)
        max_element= max(nums)

        self.freq= [0 for i in range(max_element+1)]
        self.t= [-1 for i in range(max_element+ 1)]
        for i in nums:
            self.freq[i]+=1

        print('indx: ', [i for i in range(max_element+1)])
        print('freq: ', self.freq)
        return self.delete_and_earn(max_element)


s= Solution()
```

```python
nums = [3,4,2]
# Output: 6
s.deleteAndEarn(nums)
```

```python
nums = [2,2,3,3,3,4]
# Output: 9
s.deleteAndEarn(nums)
```

### Bottom Up

```python
class Solution:
    def delete_and_earn(self, n):
        # Base Condition
        self.t[0]= 0

        for i in range(1, n+1):
            delete_i= i*self.freq[i]
            if i>1:
                delete_i+=self.t[i-2]
            not_delete_i= self.t[i-1]
            self.t[i]= max(delete_i, not_delete_i)
        print('s.t:  ', self.t)
        return self.t[n]
        
    def deleteAndEarn(self, nums: list[int]) -> int:
        n= len(nums)
        max_element= max(nums)

        self.freq= [0 for i in range(max_element+1)]
        self.t= [-1 for i in range(max_element+ 1)]
        for i in nums:
            self.freq[i]+=1

        print('indx: ', [i for i in range(max_element+1)])
        print('freq: ', self.freq)
        return self.delete_and_earn(max_element)


s= Solution()
```

```python
nums = [3,4,2]
# Output: 6
s.deleteAndEarn(nums)
```

```python
nums = [2,2,3,3,3,4]
# Output: 9
s.deleteAndEarn(nums)
```
