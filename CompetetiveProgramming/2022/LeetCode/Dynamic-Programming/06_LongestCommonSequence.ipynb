# Recursion

```python
class Solution:
    def longestCommonSubsequence(self, text1: str, text2: str) -> int:
        n= len(text1)
        m= len(text2)

        def solve(n, m):
            if n==0 or m==0:
                return 0
            if text1[n-1]==text2[m-1]:
                return 1+ solve(n-1, m-1)
            else:
                t1_subtree= solve(n-1, m)
                t2_subtree= solve(n, m-1)
                return max(t1_subtree, t2_subtree)
        return solve(n, m)

s= Solution()
```

```python
text1 = "abcde"
text2 = "ace" 
# Output: 3  
s.longestCommonSubsequence(text1, text2)
```

```python
text1 = "abc"
text2 = "abc"
# Output: 3
s.longestCommonSubsequence(text1, text2)
```
