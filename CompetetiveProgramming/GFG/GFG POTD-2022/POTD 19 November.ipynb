## ***[Check if all levels of two trees are anagrams or not](https://practice.geeksforgeeks.org/problems/check-if-all-levels-of-two-trees-are-anagrams-or-not/1)***

```python
from typing import Optional
from collections import deque
"""

definition of binary tree node.
class Node:
    def _init_(self,val):
        self.data = val
        self.left = None
        self.right = None
"""

class Solution:
    # def areAnagrams(self, node1 : Optional['Node'], node2 : Optional['Node']) -> bool:
        # code here
     def areAnagrams(self, node1 : Optional['Node'], node2 : Optional['Node']) -> bool:

        q1=deque([[node1]])

        q2=deque([[node2]])

        

        while q1[0] and q2[0]:

           

            curr1=q1.popleft()

            curr2=q2.popleft()

            

            

            arr1=[]

            arr2=[]

            dic1={}

            dic2={}

            #print(curr1)

            for i in curr1:

             

             

                if i.left:

                    arr1.append(i.left)

                if i.right:

                    arr1.append(i.right)

                    

                if i.data in dic1:

                    dic1[i.data]+=1

                else:

                    dic1[i.data]=1

                    

            for i in curr2:

             

                if i.left:

                    arr2.append(i.left)

                if i.right:

                    arr2.append(i.right)

                    

                if i.data in dic2:

                    dic2[i.data]+=1

                else:

                    dic2[i.data]=1

                    

            if dic1!=dic2:

                return False

                

            q1.append(arr1)

            q2.append(arr2)

            

        if q1[0] or q2[0]:

            return False

        return True
        


```
