```python
#User function Template for python3

class Solution:
    def maxGroupSize(self, arr, N, K):
        # code here 
        maxlen= 0
        for i in range(N):
            l= [] # empty list
            for a in arr[i:]:
                if len(l)==0:
                    l.append(a)
                else:
                    for j in l:
                        if (a+j)%K==0:
                            print(f'reached {a}, BREAK')
                            break
                    print(f'appended: {a}')
                    l.append(a)
            maxlen= max(maxlen, len(l))
            print(f'i= {i}, l= {l}')

                
        return maxlen


s= Solution()
```

```python
N = 4
K = 8
arr = [1, 7, 2, 6]


s.maxGroupSize(arr, N, K)
```

```python
class Solution:
    def maxGroupSize(self, arr, N, K):
        # code here 
        remainder = [0]*K
        print(f'remainder(initialized)= {remainder}')
        for i in range(N):
            remainder[arr[i]%K] += 1


        print(f'remainder(modified)= {remainder}')
        ans = 0


        for i in range(1,int(K/2)+1):
            print(f'i= {i}')
            if i != K-i:
                ans += max(remainder[i],remainder[K-i])
            else:
                if remainder[i] != 0:
                    ans += 1
        if remainder[0] != 0:
            ans += 1
        
        return ans

s= Solution()
```

```python
class Solution:
    def maxGroupSize(self, arr, N, K):
        # code here 
        remainder_freq= dict()
        for i in range(K):
            remainder_freq[i]= 0
        for i in arr:
            remainder_freq[i%K]+=1
        # print(f'remainder frequency= {remainder_freq}')


        rem_keys= list(remainder_freq.keys())
        # print(f'rem_keys= {rem_keys}')
        # print(f'rem_values= {list(remainder_freq.values())}')


        i= 1
        j= rem_keys[-1]
        maxlen= 0

        while i<j:
            maxlen+= max(remainder_freq[i], remainder_freq[j])
            i+=1
            j-=1

        if i==j and remainder_freq[i]!=0:
            maxlen+=1
        if remainder_freq[0]!=0:
            maxlen+=1

        return maxlen

s= Solution()
```

```python
N = 4
K = 8
arr = [1, 7, 2, 6]


s.maxGroupSize(arr, N, K)
```
