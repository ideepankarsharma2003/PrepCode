```python
class Solution:

    def findSubarray(self,a, n):
        # code here
        res= {}
        maxsum= -10**6
        pos1= 0
        pos2= 0
        sum1= 0

        for i in range(n):
            if a[i]<0: 
                maxsum, res[sum1], sum1, pos1, pos2 = max(maxsum, sum1), res.get(sum1, [])+[a[pos1:pos2]], 0, i+1, i+1
            else: 
                sum1, pos2 = sum1+a[i], i+1   
        maxsum, res[sum1], maxlen, res1 = max(maxsum, sum1), res.get(sum1, [])+[a[pos1:pos2]], float("-inf"), []

        for i in res[maxsum]:
            if len(i)>maxlen: 
                maxlen, res1 = len(i), i

        if len(res1)==0: 
            return [-1]

        return res1

s= Solution()
```
