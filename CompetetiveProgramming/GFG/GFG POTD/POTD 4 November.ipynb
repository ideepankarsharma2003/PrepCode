```python
def int2base(num, base):
    if num == 0:
        return [0]
    digits = []
    while num:
        digits.append((int(num % base)))
        num //= base
    # print(digits)
    digits.reverse()
    digits= [str(i) for i in digits]
    return ''.join(digits)
```
