# **[House Robber](https://leetcode.com/explore/featured/card/dynamic-programming/631/strategy-for-solving-dp-problems/4148/)**

Recursion

```python
class Solution:
    def rob_house(self, arr, n):
        # base condition
        if n==1:
            return arr[0]
        if n==0:
            return 0

        house_robbed= arr[n-1]+ self.rob_house(arr, n-2)
        house_not_robbed= self.rob_house(arr, n-1)
        print(house_not_robbed, house_robbed)
        return max(house_robbed, house_not_robbed)

    def rob(self, nums) -> int:
        n= len(nums)
        return self.rob_house(nums, n)

s= Solution()
```

```python
nums = [1,2,3,1]
# Output: 4
s.rob(nums)
```

```python
nums = [2,7,9,3,1]
# Output: 12
s.rob(nums)
```

Memoization

```python
class Solution:
    def rob_house(self, arr, n):
        # base condition
        if n==1:
            return arr[0]
        if n==0:
            return 0

        if self.t[n]!=-1:
            return self.t[n]
        

        house_robbed= arr[n-1]+ self.rob_house(arr, n-2)
        house_not_robbed= self.rob_house(arr, n-1)
        print(house_not_robbed, house_robbed)
        self.t[n]= max(house_robbed, house_not_robbed)
        return self.t[n]

    def rob(self, nums) -> int:
        n= len(nums)
        self.t= [-1 for i in range(n+1)] # array for memoization
        return self.rob_house(nums, n)

s= Solution()
```

```python
nums = [1,2,3,1]
# Output: 4
s.rob(nums)
```

```python
nums = [2,7,9,3,1]
# Output: 12
s.rob(nums)
```

Bottom Up

```python
class Solution:
    def rob_house(self, arr, n):
        # base condition
        self.t[0]= 0
        self.t[1]= arr[0]

        for i in range(2, n+1):
            house_robbed= arr[i-1]+ self.t[i-2]
            house_not_robbed= self.t[i-1]
            self.t[i]= max(house_robbed, house_not_robbed)
        return self.t[n]

    def rob(self, nums) -> int:
        n= len(nums)
        self.t= [-1 for i in range(n+1)] # array for Bottom Up 
        self.t[0]= 0
        self.t[1]= nums[0]
        return self.rob_house(nums, n)

s= Solution()
```

```python
nums = [1,2,3,1]
# Output: 4
s.rob(nums)
```

```python
nums = [2,7,9,3,1]
# Output: 12
s.rob(nums)
```
