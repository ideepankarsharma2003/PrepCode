### Recursion

```python
class Solution:
    def min_cost(self, arr, n):
        if n==0:
            return 0
        if n==1:
            return arr[0]

        step_1= self.min_cost(arr, n-1)
        step_2= self.min_cost(arr, n-2)
        
        return arr[n-1]+ min(step_1, step_2)
    def minCostClimbingStairs(self, cost) -> int:
        cost.reverse()
        print(cost)
        n= len(cost)
        return min(self.min_cost(cost, n), self.min_cost(cost, n-1))
        
s= Solution()
```

```python
cost = [10,15,20]
# Output: 15
s.minCostClimbingStairs(cost)
```

### Memoization

```python
class Solution:
    def min_cost(self, arr, n):
        if n==0:
            return 0
        if n==1:
            return arr[0]

        if self.t[n]!=-1:
            return self.t[n]

        step_1= self.min_cost(arr, n-1)
        step_2= self.min_cost(arr, n-2)
        
        self.t[n]= arr[n-1]+ min(step_1, step_2)
        return self.t[n]

    def minCostClimbingStairs(self, cost) -> int:
        cost.reverse()
        print(cost)
        n= len(cost)
        self.t= [-1 for i in range(n+1)]
        return min(self.min_cost(cost, n), self.min_cost(cost, n-1))
        
s= Solution()
```
