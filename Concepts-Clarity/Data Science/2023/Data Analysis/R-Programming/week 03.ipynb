```r
penguins %>% arrange(bill_length_mm) # sort in ascending order
```

```r
penguins %>% arrange(-bill_length_mm) # sort in descending order
```

```r
penguins2 <- penguins %>% arrange(-bill_length_mm) # sort in descending order
View(penguins2)
```

```r
penguins %>%
    group_by(island) %>%
    drop_na() %>%
    summarise(mean_bill_length_mm = mean(bill_length_mm))

```

```r
penguins %>%
    group_by(island) %>%
    drop_na() %>%
    summarise(max_bill_length_mm = max(bill_length_mm))
```

```r
penguins %>%
    group_by(species, island) %>%
    drop_na() %>%
    summarise(mean_bill_length_mm = mean(bill_length_mm), max_bill_length_mm = max(bill_length_mm))
```
