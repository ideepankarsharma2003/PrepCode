# **[Outlier Detection and Removal using Percentile](https://github.com/codebasics/py/tree/master/ML/FeatureEngineering/1_outliers)**

```python
import pandas as pd
```

```python
df= pd.read_csv('heights.csv')
df.head()
```

```python
df['height(in feets)']
```

```python
df['height(in feets)'].quantile(0.90)
```

```python
max_threshold= df['height(in feets)'].quantile(.95)
max_threshold
```

```python
df[df['height(in feets)']>max_threshold]
```

```python
min_threshold= df['height(in feets)'].quantile(.05)
min_threshold
```

```python
df[df['height(in feets)']<min_threshold]
```

```python
df[(df['height(in feets)']<max_threshold ) &  (df['height(in feets)']>min_threshold)]
```

## **Banglore House Prediction**

```python
df= pd.read_csv('bhp.csv')
df.head()
```

```python
df.describe()
```

```python
min_threshold, max_threshold= df.price_per_sqft.quantile((0.01, .99))
min_threshold, max_threshold
```

```python
df[df['price_per_sqft']<min_threshold]
```

```python
df[df['price_per_sqft']>max_threshold]
```

```python
df2= df[(df['price_per_sqft']>min_threshold )& (df['price_per_sqft']<max_threshold)]
df2.sample(10)
```
