# **Measure of Central Tendency**

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data= pd.read_csv('iris.data.csv')
data.sample(5)
```

```python
data.head()
```

```python
data.columns = ["sl", "sw", "pl", "pw", "species"]
```

```python
data.sample(10)
```

```python
data.species.value_counts()
```

```python
iris_setosa= data[data.species=='Iris-setosa']
iris_virginica= data[data.species=='Iris-virginica']
iris_versicolor= data[data.species=='Iris-versicolor']
```

## Mean <br>


```python
data.columns
```

```python
# sepal-width
print(np.mean(iris_setosa.sw))
print(np.mean(iris_versicolor.sw))
print(np.mean(iris_virginica.sw))
```

```python
# sepal-length
print(np.mean(iris_setosa.sl))
print(np.mean(iris_versicolor.sl))
print(np.mean(iris_virginica.sl))
```

```python
# petal-width
print(np.mean(iris_setosa.pw))
print(np.mean(iris_versicolor.pw))
print(np.mean(iris_virginica.pw))
```

```python
# petal-length
print(np.mean(iris_setosa.pl))
print(np.mean(iris_versicolor.pl))
print(np.mean(iris_virginica.pl))
```

### The problem of outliers

```python
print(np.max(iris_setosa.pl))
print(np.min(iris_setosa.pl))
```

```python
# range of setosa petal length is b/w (1-2)
plt.hist(iris_setosa.pl)
```

```python
# Mean ---> suffers from outliers
print(f'original mean: {np.mean(iris_setosa.pl)}')
print(f'mutated mean: {np.mean(np.append(iris_setosa.pl, 50))}') # bas 1 outlier 50 append kiya
```

# Median <br>


```python
# petal-length
print(np.median(iris_setosa.pl))
print(np.median(iris_versicolor.pl))
print(np.median(iris_virginica.pl))
```

```python
# Good thing about median ---> not much impact of outliers
# Jab tak aaadhe se jyada values outlier nhi ho jati, Median remains unaffected
# petal-length
print(np.median(np.append(iris_setosa.pl, 50)))
print(np.median(np.append(iris_versicolor.pl, 50)))
print(np.median(np.append(iris_virginica.pl, 50)))
```

# Mode<br>


```python
from scipy import stats

# petal-length
print(stats.mode(iris_setosa.pl))
print(stats.mode(iris_versicolor.pl))
print(stats.mode(iris_virginica.pl))
```
