# **Linked List**<br/>


````python
class Node:
    def __init__(self, data=None, next=None) -> None:
        self.data= data
        self.next= next
````

````python
class LinkedList:
    def __init__(self) -> None:
        self.head= None
        
    def insertAtBeginning(self, data): # insert the data at the beginning of the node
        node= Node(data, self.head)
        self.head= node

    def insertAtEnd(self, data): # insert the data at the End of the node
        node= Node(data)
        if self.head is None:
            self.head= node
            return
        itr= self.head
        while itr.next:
            itr= itr.next
        itr.next= node

    def removeAtIndex(self, index): # remove the element at index
        if index<0 or index>self.getLength():
            print("Invalid Index")
            return
        if index==0:
            self.head= self.head.next
            return
        
        else:
            itr= self.head
            count= 0
            while itr:
                if count==index-1:
                    itr.next= itr.next.next
                    return
                itr= itr.next
                count+=1

    def insertAtIndex(self, index, data): # insert the element at index
        if index<0 or index>self.getLength():
            print("Invalid Index")
            return
        if index==0:
            self.insertAtBeginning(data)
            return
        
        else:
            itr= self.head
            count= 0
            while itr:
                if count==index-1:
                    itr.next= Node(data, itr.next)
                    return
                itr= itr.next
                count+=1
    
    
    def insertValues(self, data_list): # insert a list of values
        for i in data_list:
            self.insertAtEnd(i)

    def emptyList(self): # empty the Linked List
        self.head= None   

    def getLength(self): # get the length of the Linked List
        count= 0
        itr= self.head
        while itr:
            count+=1
            itr= itr.next
        return count


    def printLinkedList(self): # print the list 
        if self.head==None:
            print("The List is empty !!!")
            return
        
        else:
            linklist= ''
            itr= self.head
            while itr!=None:
                linklist+= str(itr.data)+ ' ----> '
                itr= itr.next
            linklist+='None'
            print(linklist)


    '''Exercise'''
    def insert_after_value(self, data_after, data_to_insert):
        # Search for first occurance of data_after value in linked list
        # Now insert data_to_insert after data_after node
        itr= self.head
        while itr:
            if itr.data==data_after:
                itr.next= Node(data_to_insert, itr.next)
                return
            itr= itr.next
        print(f'{data_after} does not exist in linked list')

    def remove_by_value(self, data):
        # Remove first node that contains data
        itr= self.head
        count= 0
        while itr:
            if itr.data==data:
                self.removeAtIndex(count)
                return
            itr= itr.next
            count+=1
        print(f'{data} does not exist in linked list')


````

# **Doubly Linked List**<br/>


````python
class Node:
    def __init__(self, data=None, prev= None, next= None) -> None:
        self.data= data
        self.prev= prev
        self.next= next
````

````python
class DoublyLinkedList:
    def __init__(self) -> None:
        self.head= None
        
    def insertAtBeginning(self, data): # insert the data at the beginning of the node
        prevHead= self.head
        node= Node(data, None, self.head)

        self.head= node
        if node.next:
            prevHead.prev= node
        

    def insertAtEnd(self, data): # insert the data at the End of the node
        node= Node(data)
        if self.head is None:
            self.head= node
            return
        itr= self.head
        while itr.next:
            itr= itr.next
        node.prev= itr
        itr.next= node

    def getLastElement(self): # get last element of the list
        itr= self.head
        while itr.next:
            itr= itr.next
        return itr
    
    def insertValues(self, data_list): # insert a list of values
        for i in data_list:
            self.insertAtEnd(i)



    def removeAtIndex(self, index): # remove the element at index
        if index<0 or index>self.getLength():
            print("Invalid Index")
            return
        if index==0:
            self.head= self.head.next
            self.head.prev= None
            return
        
        else:
            itr= self.head
            count= 0
            while itr:
                if count==index-1:
                    itr.next= itr.next.next
                    itr.next.prev= itr
                    return
                itr= itr.next
                count+=1

    def insertAtIndex(self, index, data): # insert the element at index
        if index<0 or index>self.getLength():
            print("Invalid Index")
            return
        if index==0:
            self.insertAtBeginning(data)
            return
        
        else:
            itr= self.head
            count= 0
            while itr:
                if count==index-1:
                    itr.next= Node(data, itr, itr.next)
                    if itr.next.next:
                        itr.next.next.prev= itr.next
                    # itr.next.prev= itr
                    return
                itr= itr.next
                count+=1


    def emptyList(self): # empty the Linked List
        self.head= None
        self.prev= None   

    def getLength(self): # get the length of the Linked List
        count= 0
        itr= self.head
        while itr:
            count+=1
            itr= itr.next
        return count
        
    def printDoublyLinkedListForward(self): # print the list forward
        if self.head==None:
            print("The List is empty !!!")
            return
        
        else:
            linklist= 'None <----> '
            itr= self.head
            while itr!=None:
                linklist+= str(itr.data)+ ' <----> '
                itr= itr.next
            linklist+='None'
            print(linklist)

    def printDoublyLinkedListBackward(self): # print the list backward
        if self.head==None:
            print("The List is empty !!!")
            return
        
        else:
            tail= self.getLastElement()
            linklist= 'None <----> '
            itr= tail
            while itr!=None:
                linklist+= str(itr.data)+ ' <----> '
                itr= itr.prev
            linklist+='None'
            print(linklist)

    
````

````python
class HashTable:
    def __init__(self) -> None:
        self.MAX= 10
        self.arr= [None for i in range(self.MAX)]

    def getHash(self, key):
        sum= 0
        for char in key:
            sum+=ord(char)
        return sum%self.MAX

    # def add(self, key, val):
    def __setitem__(self, key, val):
        h= self.getHash(key)
        self.arr[h]= val
    
    # def get(self, key):
    def __getitem__(self, key):
        h= self.getHash(key)
        return self.arr[h]
    
    def __delitem__(self, key):
        h= self.getHash(key)
        self.arr[h]= None
        
````

## **Chaining**


````python
class HashTableChaining:
    def __init__(self) -> None:
        self.MAX = 10
        self.arr = [[] for i in range(self.MAX)]

    def getHash(self, key):
        sum = 0
        for char in key:
            sum += ord(char)
        return sum % self.MAX

    # def add(self, key, val):
    def __setitem__(self, key, val):
        h = self.getHash(key)
        # found= False # check if the key already exists
        for idx, element in enumerate(self.arr[h]): 
            # check if the key already exists
            if len(element) == 2 and key == element[0]:
                self.arr[h][idx]= (key, val)
                # found= True
                return
        self.arr[h].append((key, val))
            


    # def get(self, key):
    def __getitem__(self, key):
        h = self.getHash(key)
        # self.arr[h] is a list
        # element is a tuple (key, value)
        for element in self.arr[h]: 
            if element[0]==key:
                return element[1]


    def __delitem__(self, key):
        h = self.getHash(key)
        # self.arr[h] = None
        for idx, element in enumerate(self.arr[h]):
            if element[0]==key:
                del self.arr[h][idx]
````

````python
class HashTableLinearProbing:
    def __init__(self) -> None:
        self.MAX = 10
        self.arr = [None for i in range(self.MAX)]

    def getHash(self, key):
        sum = 0
        for char in key:
            sum += ord(char)
        return sum % self.MAX

    # def add(self, key, val):
    def __setitem__(self, key, val):
        h = self.getHash(key)
        count= 0
        if self.arr[h] is None:
            self.arr[h]= (key, val, count, h)
            return
        else:
            for i in range(h, self.MAX):
                if self.arr[i] is None:
                    self.arr[i] = (key, val, count, h)
                    return
                count+=1
            for i in range(0, h):
                if self.arr[i] is None:
                    self.arr[i] = (key, val, count, h)
                    return
                count+=1
        print("No more space left")
                    




    # def get(self, key):
    def __getitem__(self, key):
        h = self.getHash(key)

        count=0
        if self.arr is not None and self.arr[h][3]==h and self.arr[h][0]==key:
            return self.arr[h][1]
        else:
            for i in range(h, self.MAX):
                # print(self.arr[i])
                if self.arr[i] is None:
                    return
                if self.arr[i][3] == h and self.arr[i][0] == key:
                    return self.arr[i][1]
                count += 1
            for i in range(0, h):
                # print(self.arr[i])
                if self.arr[i] is None:
                    return
                if self.arr[i][3] == h and self.arr[i][0] == key:
                    return self.arr[i][1]
                count += 1
            return 'not in hash table'


    def __delitem__(self, key):
        h = self.getHash(key)

        count=0 

        if self.arr[h] is not None and key==self.arr[h][0]:
            del self.arr[h]
        else:
            for i in range(h, self.MAX):
                if self.arr[i] is not None and self.arr[i][3] == h and self.arr[i][0] == key:
                    del self.arr[i]
                count += 1
            for i in range(0, h):
                # print(self.arr[i])
                if self.arr[i] is not None and self.arr[i][3] == h and self.arr[i][0] == key:
                    del self.arr[i]
                count += 1
            return 'not in hash table'
````

````python
# creating an own stack class
class Stack:
    def __init__(self) -> None:
        self.node= deque() # making a node using deque 
    def push(self, val):
        self.node.append(val)
    def pop(self):
        return self.node.pop()
    def peek(self):
        return self.node[-1]
    def isEmpty(self):
        return len(self.node)==0
    def size(self):
        return len(self.node)
    
````

````python
def reverse_string(s):
    stk= Stack()
    ans= ''
    for i in s:
        stk.push(i)
    while not stk.isEmpty():
        ans+=stk.pop()
    return ans
    
````

````python
def is_balanced(s):
    stk= Stack()
    for i in s:
        # print(f'stack= {stk.node}')
        if i=='(' or i=='{' or i=='[':
            stk.push(i)
        elif i==')' or i=='}' or i==']':
            if stk.isEmpty():
                return False
            stk.pop()
    return stk.isEmpty()
````

````python
# Queue using collections.deque
from collections import deque
q= deque()
for i in range(10, 21):
    q.appendleft(i) # insert in queue, at index 0
print(f'queue= {q}')
print('popped: ', q.pop()) # pop the first element from the queue, FIFO order
print(f'queue= {q}')
````

````python
# creating own Queue class
class Queue:
    def __init__(self) -> None:
        self.node= deque()
    def enqueue(self, val) :
        self.node.appendleft(val)
    def dequeue(self) :
        if self.isEmpty():
            return None
        return self.node.pop()
    def size(self) :
        return len(self.node)
    def isEmpty(self) :
        return self.size()==0
    def front(self) : # return the first element of the queue, FIFO
        return self.node[-1]
````

````python
q= Queue()
print(f'isEmpty= {q.isEmpty()}')
for i in range(11, 21):
    q.enqueue(i) # insert in queue, at index 0
print(f'queue= {q.node}')
print('popped: ', q.dequeue()) # pop the first element from the queue, FIFO order
print(f'queue= {q.node}')
print(f'queue size= {q.size()}')
print(f'isEmpty= {q.isEmpty()}')
````

````python
class TreeNode:
    def __init__(self, data) -> None:
        self.data= data
        self.children= []
        self.parent= None

    def addChild(self, child):
        child.parent= self
        self.children.append(child)

    def getLevel(self): # return the level of the node
        level= 0
        p= self.parent
        while p:
            level+=1
            p= p.parent
        return level

    def printTree(self):
        spaces= 4* ' '* self.getLevel()
        prefix= spaces+ '|--'
        print(prefix+ self.data)
        if self.children:
            for child in self.children:
                child.printTree()
````

````python
class TreeNodeCompanyHeirarchy:
    def __init__(self, name, designation) -> None:
        self.name = name
        self.designation = designation
        self.children = []
        self.parent = None

    def addChild(self, child):
        child.parent = self
        self.children.append(child)

    def getLevel(self):  # return the level of the node
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level

    def printTree(self, query):
        if query=='name':
            self.printNameTree()
        elif query=='designation':
            self.printDesignationTree()
        elif query=='both':
            self.printBothTree()
        

    def printNameTree(self):
        spaces = 4 * ' ' * self.getLevel()
        prefix = spaces + '|--'
        print(prefix + self.name)
        if self.children:
            for child in self.children:
                child.printNameTree()

    def printDesignationTree(self):
        spaces = 4 * ' ' * self.getLevel()
        prefix = spaces + '|--'
        print(prefix + self.designation)
        if self.children:
            for child in self.children:
                child.printDesignationTree()

    def printBothTree(self):
        spaces = 4 * ' ' * self.getLevel()
        prefix = spaces + '|--'
        print(f'{prefix}{self.name} ({self.designation})')
        if self.children:
            for child in self.children:
                child.printBothTree()
````

````python
class TreeNodePlaces:
    def __init__(self, name) -> None:
        self.name = name
        self.children = []
        self.parent = None

    def addChild(self, child):
        child.parent = self
        self.children.append(child)

    def getLevel(self):  # return the level of the node
        level = 0
        p = self.parent
        while p:
            level += 1
            p = p.parent
        return level



    
    def printTreeUptoLevel(self, level):
        if self.getLevel()<=level:
            spaces = 4 * ' ' * self.getLevel()
            prefix = spaces + '|--'
            print(prefix + self.name)
            if self.children:
                for child in self.children:
                    child.printTreeUptoLevel(level)
    
    def printTree(self):
        spaces = 4 * ' ' * self.getLevel()
        prefix = spaces + '|--'
        print(prefix + self.name)
        if self.children:
            for child in self.children:
                child.printTree()

    
````
