# **Metrics in sklearn**


```python
from sklearn import linear_model
import pandas as pd
import numpy as np
```

```python
path= 'D:/Placement-Prep/Concepts-Clarity/Data Science/ML_Sololearn/DataSets/titanic.csv'
df = pd.read_csv(path)
df
```

```python
df['male'] = df['Sex'] == 'male'
df
```

```python
X = df[['Pclass', 'male', 'Age', 'Siblings/Spouses', 'Parents/Children', 'Fare']].values
y = df['Survived'].values
model = linear_model.LogisticRegression()
model.fit(X, y)
y_pred = model.predict(X)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print(f'accuracy_score: {accuracy_score(y, y_pred)} ')
print(f'precision_score: {precision_score(y, y_pred)} ')
print(f'recall_score: {recall_score(y, y_pred)} ')
print(f'f1_score: {f1_score(y, y_pred)} ')
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_pred, y))
```
