# **[Knapsack with Duplicate Items](https://practice.geeksforgeeks.org/problems/knapsack-with-duplicate-items4201/1)**

### **Recursion**

```python
class Solution:
    def knapSack(self, N, W, val, wt):
        # code here
        if N==0 or W==0:
            return 0
        if wt[N-1]<=W:
            current_value= val[N-1]
            current_value_included= current_value+self.knapSack(N, W-wt[N-1], val, wt)
            current_value_excluded= self.knapSack(N-1, W, val, wt)
            return max(current_value_included, current_value_excluded)
        else:
            current_value_excluded= self.knapSack(N-1, W, val, wt)
            return current_value_excluded

s= Solution()
```

```python
N = 2
W = 3
val= [1, 1]
wt= [2, 1]
s.knapSack(N, W, val, wt)
```

```python
N = 4
W = 8
val= [1, 4, 5, 7]
wt= [1, 3, 4, 5]
s.knapSack(N, W, val, wt)
```

### **Memoization**

`Stack Overflow -.-`

```python
class Solution:
    def knapSack(self, N, W, val, wt, init=True):
        # code here
        if init:
            self.t= [[False for i in range(W+1)] for i in range(N+1)]
            init=False
        if N==0 or W==0:
            return 0

        if self.t[N][W]:
            return self.t[N][W]

        if wt[N-1]<=W:
            current_value= val[N-1]
            current_value_included= current_value+self.knapSack(N, W-wt[N-1], val, wt, init)
            current_value_excluded= self.knapSack(N-1, W, val, wt, init)
            self.t[N][W]= max(current_value_included, current_value_excluded)
            return self.t[N][W]
        else:
            current_value_excluded= self.knapSack(N-1, W, val, wt, init)
            self.t[N][W]= current_value_excluded
            return self.t[N][W]

s= Solution()
```

```python
N = 2
W = 3
val= [1, 1]
wt= [2, 1]
s.knapSack(N, W, val, wt)
```

```python
N = 4
W = 8
val= [1, 4, 5, 7]
wt= [1, 3, 4, 5]
s.knapSack(N, W, val, wt)
```

### **Bottom Up**

```python
class Solution:
    def knapSack(self, N, W, val, wt):
        # code here
        self.t= [[0 for i in range(W+1)] for i in range(N+1)]

        for i in range(N+1):
            for j in range(W+1):
                if wt[i-1]<=j:
                    current_value= val[i-1]
                    current_value_included= current_value+self.t[i][j-wt[i-1]]
                    current_value_excluded= self.t[i-1][j]
                    self.t[i][j]= max(current_value_included, current_value_excluded)
                else:
                    current_value_excluded= self.t[i-1][j]
                    self.t[i][j]= (current_value_excluded)

        return self.t[-1][-1]


        

        

s= Solution()
```
