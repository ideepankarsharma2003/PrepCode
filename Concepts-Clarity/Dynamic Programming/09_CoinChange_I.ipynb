# **[322. Coin Change](https://leetcode.com/problems/coin-change/)**

```python
class Solution:
    def coinChange(self, coins, amount: int) -> int:
        n= len(coins)
        INTMAX= 2**32 # constraints
        self.t= [[INTMAX for i in range(amount+1)] for i in range(n+1)]
        for i in range(n+1):
            self.t[i][0]=0

        for j in range(1, amount+1):
            if j%coins[0]==0:
                self.t[1][j]= j/coins[0]

        for i in range(1, n+1):
            for j in range(1, amount+1):
                if coins[i-1]<=j:
                    self.t[i][j]= min(1+self.t[i][j-coins[i-1]], self.t[i-1][j])
                else:
                    self.t[i][j]= self.t[i-1][j]
        if self.t[-1][-1]==INTMAX:
            return -1
        return self.t[-1][-1]

s= Solution()
```

```python
coins = [1,2,5]
amount = 11
s.coinChange(coins, amount)
```
