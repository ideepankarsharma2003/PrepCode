# **[518. Coin Change II](https://leetcode.com/problems/coin-change-ii/)**

```python
class Solution:
    def change(self, amount, coins) -> int:
        n= len(coins)
        self.t= [[0 for i in range(amount+1)]for i in range(n+1)]
        for i in range(n+1):
            self.t[i][0]=1
        
        for i in range(1, n+1):
            for j in range(amount+1):
                if coins[i-1]<=j:
                    self.t[i][j]= self.t[i][j-coins[i-1]] + self.t[i-1][j]
                else:
                    self.t[i][j]= self.t[i-1][j]
        
        return self.t[-1][-1]

s= Solution()
```

```python
amount = 5
coins = [1,2,5]
s.change(amount, coins)
```

```python
amount = 10
coins = [10]
s.change(amount, coins)
```
