## **Recursion**

```python

class Solution:
    #Function to find the length of longest common subsequence in two strings.
    def lcs(self,x,y,s1,s2):
        # code here
        if x==0 or y==0:
            return 0        
        if s1[x-1]==s2[y-1]:
            return 1+ self.lcs(x-1, y-1, s1, s2)
        else:
            return max(self.lcs(x-1, y, s1, s2), self.lcs(x, y-1, s1, s2))
s= Solution()
```

```python
A = 6
B = 6
str1 = 'ABCDGH'
str2 = 'AEDFHR'
s.lcs(A, B, str1, str2)
```

```python
A = 3
B = 2
str1 = 'ABC'
str2 = 'AC'
s.lcs(A, B, str1, str2)
```

## **Memoization**

```python

class Solution:
    #Function to find the length of longest common subsequence in two strings.
    def lcs(self,x,y,s1,s2, temp=True):
        # code here
        if temp:
            self.t= [[False for i in range(y+1)]for j in range(x+1)]
            # for i in self.t:
            #     print(i)
            temp= False
            
        if x==0 or y==0:
            return 0      
        # print(x, y)
        if self.t[x][y]:
            return self.t[x][y]  

        if s1[x-1]==s2[y-1]:
            self.t[x][y]= 1+ self.lcs(x-1, y-1, s1, s2, temp)
            return self.t[x][y]
        else:
            self.t[x][y]= max(self.lcs(x-1, y, s1, s2, temp), self.lcs(x, y-1, s1, s2, temp))
            return self.t[x][y]
s= Solution()
```
