## **Recursion**

```python

class Solution:
    #Function to find the length of longest common subsequence in two strings.
    def longestCommonSubstr(self, S1, S2, n, m, continuous=0):
        # code here
        if n==0 or m==0:
            return continuous
        if S1[n-1]==S2[m-1]:
            continuous= self.longestCommonSubstr(S1, S2, n-1, m-1, continuous+1)
        
        a= self.longestCommonSubstr(S1, S2, n-1, m, 0) # break continuity , reset to 0    
        b= self.longestCommonSubstr(S1, S2, n, m-1, 0) # break continuity , reset to 0
        return max(a, b, continuous)
s= Solution()
```

```python
S1 = "ABC"
S2 = "ACB"
n = 3
m = 3
s.longestCommonSubstr(S1, S2, n, m)
```
